```python
import torch

# Load the YOLOv5 model with the 'best.pt' weights
model = torch.hub.load('ultralytics/yolov5', 'custom', path=r"########", force_reload=True)
```

```python
# Load an example image for prediction
image_path = r"#####"  # Replace with the path to your input image

# Perform inference on the input image
results = model(image_path)

# Get the predicted bounding boxes, labels, and confidence scores
predicted_boxes = results.pred[0][:, :4].cpu().numpy()  # Bounding boxes in (x_min, y_min, x_max, y_max) format
predicted_scores = results.pred[0][:, 4].cpu().numpy()   # Confidence scores for each detection
predicted_labels = results.pred[0][:, 5].cpu().numpy().astype(int)  # Integer labels for each detection

# You can also get the class names from the model
class_names = model.names
predicted_class_names = [class_names[label] for label in predicted_labels]

# Print the predictions
for box, score, label in zip(predicted_boxes, predicted_scores, predicted_class_names):
    print(f"Label: {label}, Confidence: {score:.2f}, Bounding Box: {box}")
```

```python
from PIL import Image, ImageDraw, ImageFont

image = Image.open(image_path)
for box, score, label in zip(predicted_boxes, predicted_scores, predicted_labels):
    label_str = model.names[label]
    box = [int(coord) for coord in box]

    # Draw the bounding box
    draw = ImageDraw.Draw(image)
    draw.rectangle(box, outline='red', width=10)
    
    # Increase the text size by using a larger image size
    font_size = 50  # Set a larger font size
    font = ImageFont.truetype("arial.ttf", font_size)

    # Draw the class label and confidence
    text = f"{label_str}: {score:.2f}"
    draw.text((box[0], box[1] - 60), text, fill='red', font = font)

# Show the image with bounding boxes
image.show()
```
